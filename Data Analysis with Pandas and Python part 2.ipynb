# The .apply() Method

```python
def classify_performance(number):
    if number<300:
        return "OK"
    elif number>= 300 and number < 650:
        return "Satisfactory"
    else:
        return "Incredible"
```

```python
google.apply(classify_performance).tail()
```

```python
google.apply(lambda stock_price: stock_price + 1)
```

```python
nba["Sport"] = "Basketball"
nba.head(3)
```

```python
# substract 20,000 for each value in salary

nba["Salary"].sub(20000)
```

```python
# multiply each value in weight col by .25

nba["Weight"]*0.25
```
